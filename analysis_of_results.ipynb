# Import of modules and import of txt files

```python
import pandas as pd
import pickle
import glob
from os.path import basename
```

```python
filepaths_translated = []

for name in glob.glob('C:/Users/alexa/OneDrive/Documenten/Thesis Lab/Legal Documents/FRA_Translated/*.txt'):
    filepaths_translated.append(name)
```

```python
filepaths_original = []

for name in glob.glob('C:/Users/alexa/OneDrive/Documenten/Thesis Lab/Legal Documents/FRA_Preprocessed/*.txt'):
    filepaths_original.append(name)
```

# Creation of 4 lists: filenames (using remove_dir function), countrycode (input variable open), translated sentences and original sentences

```python
def remove_dir(filepath):
    return basename(filepath)
```

```python
sentences_translated = []
filenames = []


for filepath in filepaths_translated:
    with open(filepath, encoding='utf-8') as file:
        for line in file:   
            sentences_translated.append(line)
            filenames.append(remove_dir(filepath))
```

```python
sentences_original = []
countrycode = []

for filepath in filepaths_original:
    with open(filepath, encoding='utf-8') as file:
        for line in file:
            sentences_original.append(line)
            countrycode.append('FRA')
            
```

```python
filenames = pd.DataFrame(filenames, columns = ['Filename'])
```

```python
sentences_translated = pd.DataFrame(sentences_translated, columns = ['Sentence_translated'])
```

```python
sentences_original = pd.DataFrame(sentences_original, columns = ['Sentence_original'])
```

```python
countrycode = pd.DataFrame(countrycode, columns = ['countrycode'])
```

# Import of saved pickle model (Linear SVM B.O.W. Weighted) and application to translated_sentences. 

```python
pickled_model = pickle.load(open('C:/Users/alexa/OneDrive/Documenten/Thesis Lab/Models/tfidf_bow_translated_weighted.pkl', 'rb'))
```

```python
arr = pickled_model.predict(sentences_translated['Sentence_translated'])
```

```python
prediction = pd.DataFrame(arr, columns =['Label'])
```

```python
prediction
```

# Concatenation of 4 lists and prediction for each country seperately

```python
df = pd.concat([filenames, countrycode, prediction, sentences_translated, sentences_original], axis=1, join='inner')
```
