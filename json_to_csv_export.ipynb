```python
list0 = []

for n in range(0, 32):
     list0.append(data[n]['documentName'])
        
print(list0)
```

```python
list1 = []
list2 = []
list3 = []

for n in range(0, 32):
    for i in data[n]['tokens']:
        list1.append(i['text'])
        list2.append(i['entityLabel'])
        list3.append(list0[n])
    
df = pd.DataFrame(list1, columns=['Sentence'])
df.insert(loc=0, column='Label', value=list2)
df.insert(loc=0, column='Filename', value=list3)
df.head(50)
```
